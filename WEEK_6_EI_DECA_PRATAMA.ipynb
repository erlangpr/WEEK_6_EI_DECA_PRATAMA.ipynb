### 2.A

Buatlah fungsi **hitung_rata_rata** yang menerima sebuah list angka dan mengembalikan nilai rata-rata dari angka-angka tersebut.
```
Contoh penggunaan:
angka_list = [10, 20, 30, 40, 50]
print(hitung_rata_rata(angka_list))  #Output: 30.0
```

````python
def hitung_rata_rata(angka_list):
    if not angka_list:
        return 0
    return sum(angka_list) / len(angka_list)

angka_list = [10, 20, 30, 40, 50]
print(hitung_rata_rata(angka_list))
````

### 2.B

Buatlah sebuah array NumPy dan lakukan operasi
penjumlahan, pengurangan, perkalian, dan pembagian dengan array lain.

````python
array1 = np.array([1, 2, 3, 4, 5])
array2 = np.array([1, 3, 4, 5, 6])

penjumlahan = array1 + array2
print("Penjumlahan:", penjumlahan)

pengurangan = array1 - array2
print("Pengurangan:", pengurangan)

perkalian = array1 * array2
print("Perkalian:", perkalian)

pembagian = array1 / array2
print("Pembagian:", pembagian)
````

### 2.C
Buatlah fungsi cek_genap yang menerima sebuah angka dan mengembalikan True jika angka tersebut genap, dan False jika angka tersebut ganjil.
```
Contoh penggunaan:
print(cek_genap (18)) #Output: True
print(cek_genap (7)) #Output: False
```

````python
def cek_genap(angka):
    return angka % 2 == 0

print(f"{18}: {cek_genap(18)}")  # Output: True
print(f"{7}: {cek_genap(7)}")  # Output: False
````
